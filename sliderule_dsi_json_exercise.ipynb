## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

#### Find the 10 countries with most projects

```python
# Reading world_bank_projects.json into a dataframe
x = pd.read_json('world_bank_projects.json')
x
```

```python
# Checking the count of NAN values in every column
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    print(x.isnull().sum())
```

```python
#Here I verify if any of the project names are duplicated. And hence I can group the the countries by the project names.
x.project_name.duplicated().sum()
```

```python
# Checking the number of rows and columns in the dataframe 
x.shape
```

```python
#List unique values in the df['name'] column
x.project_name.unique()
```

```python
# Grouping all countries by the number of unique projects they have
df = x.groupby('countryname')['project_name'].nunique()
```

```python
#Checking the type of df
type(df)
```

```python

# Converting df (which is of Series type) to Pandas Dataframe type
df1=pd.DataFrame(df)
# Renaming the Column Names
df1.columns = ['Project Count']
```

```python
# Sorting the dataframe by the project count
df1=df1.sort_values(by=['Project Count'],ascending=False, na_position='first')
df1.head(10)
```
